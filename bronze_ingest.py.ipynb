```python
df = spark.read.csv("/databricks-datasets/samples/population-vs-price/data_geo.csv", header=True)

# Sanitize column names by replacing spaces with underscores
for col in df.columns:
    df = df.withColumnRenamed(col, col.replace(" ", "_"))

df.write.format("delta").mode("overwrite").saveAsTable("sales_bronze")

# Commit the changes using a Delta table OPTIMIZE command
spark.sql("OPTIMIZE sales_bronze")
```
